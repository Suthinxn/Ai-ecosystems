## 3. Analyze total sales (Total) by day and by month

```python
data.sort_values(by='Date', inplace=True)
data
```

```python
data.head(10)
```

```python
result = data.groupby('Date').sum(numeric_only=True).reset_index()
result
```

```python
plt.figure(figsize=(24, 6)) 

sns.lineplot(data=result, x='Date', y='Total', marker='o', color='royalblue', linewidth=2)

plt.title("Daily Total Sales Report", fontsize=16, fontweight='bold')
plt.xlabel("Date", fontsize=12)
plt.ylabel("Total Sales", fontsize=12)
plt.xticks(rotation=45) 
plt.grid(True, linestyle='--', alpha=0.5)


plt.show()

```

```python
data['Date'] = pd.to_datetime(data['Date'])   
data.set_index('Date', inplace=True)          
monthly_sales = data.resample('M').sum()      
monthly_sales
```

```python
sns.lineplot(x=monthly_sales.index, y=monthly_sales['Total'], marker='o')
plt.title("Monthly Total Sales Report", fontsize=16, fontweight='bold')
plt.xlabel("Month", fontsize=12)
plt.ylabel("Total Sales", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 6. Create at least three types of summary visualizations, such as:

### Bar chart

```python
region_sales = data.groupby('Region')['Total'].sum().reset_index()

plt.figure(figsize=(6,4))
plt.bar(region_sales['Region'], region_sales['Total'], color='skyblue')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

### Line plot

```python
date_sales = data.groupby('Date')['Total'].sum().reset_index()

plt.figure(figsize=(8,4))
plt.plot(date_sales['Date'], date_sales['Total'], marker='o', color='green')
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Pie chart

```python
region_sales = data.groupby('Region')['Total'].sum()

plt.figure(figsize=(6,6))
plt.pie(
    region_sales,
    labels=region_sales.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
)

plt.title('Sales Distribution by Region')
plt.axis('equal')
plt.show()

```

## Assignment 1 Questions

### 3. Which region had the highest average sales per order?

```python
order_sales = data.groupby(['Region', 'Order_ID'])['Total'].sum().reset_index()
average_sales_per_region = order_sales.groupby('Region')['Total'].mean().reset_index()
average_sales_per_region = average_sales_per_region.sort_values(by='Total', ascending=False)

average_sales_per_region
```

### 4. Is there a relationship between quantity ordered and product price?

```python
plt.figure(figsize=(8,5))
plt.scatter(data['Quantity'], data['Price'], alpha=0.6, edgecolors='w')
plt.title('Scatter Plot of Quantity vs. Price')
plt.xlabel('Quantity Ordered')
plt.ylabel('Product Price')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
